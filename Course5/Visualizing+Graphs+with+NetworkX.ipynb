# Visualizing Networks

```python

import networkx as nx
import matplotlib.pyplot as plt
""" Today, we'll be looking at a distribution network containing 51 US cities.
    Each city in this network has two node attributes, location and population.
    Edges correspond to connections that our distribution network can make between cities with the weight attribute
    corresponding to the cost of traveling over that connection. """

G=nx.read_gpickle('major_us_cities')    #it will reads the Graph from bytes data
```

```python
''' draw the graph using the default spring layout
    This will draw the graph with defaults of circular red nodes,black edges and labels. '''
plt.figure(figsize=(10,9))      #it will set the figure width and height
nx.draw_networkx(G)             #it will print graph 'G' using networkx
```

```python
# See what layouts are available in networkX
[x for x in nx.__dir__() if x.endswith('_layout')]
```

```python
# Draw the graph using the random layout
plt.figure(figsize=(10,9))
pos = nx.random_layout(G)
nx.draw_networkx(G, pos)
```

```python
# Draw the graph using the circular layout
plt.figure(figsize=(10,9))
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos)
```

```python
# Draw the graph using custom node positions
plt.figure(figsize=(10,7))

pos = nx.get_node_attributes(G, 'location') #it will gives the node attribute (ie 'location') of each node 
# like this {'Albuquerque, NM': (-106, 35),'Arlington, TX': (-97, 32)}
#If you're familiar with the locations of US cities,you may be able to now better recognize this network.
nx.draw_networkx(G, pos)
```

```python
# Draw the graph adding alpha, removing labels and axis, and softening edge color (gray)
plt.figure(figsize=(10,7))

nx.draw_networkx(G, pos, alpha=0.7, with_labels=False, edge_color='.4')       #alpha is node color strength

plt.axis('off')
plt.tight_layout();
```

```python
# Draw graph with varying node color based upon no of edges it have, node size based upon population, and edge width based upon cost of traveling over that connection 
plt.figure(figsize=(10,7))

node_color = [G.degree(v) for v in G]  #we can get a list of the degrees for each node

node_size = [0.0005*nx.get_node_attributes(G, 'population')[v] for v in G] 
#population of each node and multiply each value by a small number so that the nodes aren't too large

edge_width = [0.0015*G[u][v]['weight'] for u,v in G.edges()]
''' We can also use this comprehension to get
    the weights or the costs of each edge to be used for
    the edge widths and also multiply
    this value by a small number so that the widths are appropriate.
'''

nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=0.7, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.Blues)   #setting the color map to blue

plt.axis('off')
plt.tight_layout();  #reduce padding
```

```python
# Draw specific edges and add labels to specific nodes
plt.figure(figsize=(10,7))

node_color = [G.degree(v) for v in G]   #we can get a list of the degrees for each node
node_size = [0.0005*nx.get_node_attributes(G, 'population')[v] for v in G]  
edge_width = [0.0015*G[u][v]['weight'] for u,v in G.edges()]

nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=0.7, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.Blues)

greater_than_770 = [x for x in G.edges(data=True) if x[2]['weight']>770]
''' We can use this comprehension to find the subset of edges with the weight larger than
    770 and pass this into the edgelist parameter of draw_network_edges.
'''

nx.draw_networkx_edges(G, pos, edgelist=greater_than_770, edge_color='r', alpha=0.4, width=6)

nx.draw_networkx_labels(G, pos, labels={'Los Angeles, CA': 'LA', 'New York, NY': 'NYC'}, font_size=10, font_color='w')
''' NetworkX labels to add labels to Los Angeles and New York City.
    Looking at our network visualization by separately adding specific edges and labels,
    we can now clearly see the high cost edges as well as the locations of New York City and Los Angeles.
'''
plt.axis('off')
plt.tight_layout();
```
